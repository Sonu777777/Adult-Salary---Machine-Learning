```python
# imported all the requird libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# read the dataset by using pandas

data=pd.read_csv(r'C:\Users\Sonu\Desktop\data\Adult.csv',header=None)
data
```

```python
# assigining column name in dataset

data.columns = ["age","workclass","fnlwgt","education","education_num","marital_status","occupation","realtionship","race","sex",
                "capital_gain","capital_loss","hours_per_week","native_country","income"]
data
```

```python
# # by using pandas function checking null value present or not in dataset

data.isnull().sum()
```

```python
# droping unusefull data from the datset

data1=data.drop(['education','fnlwgt'],axis=1)
data1.shape
```

```python
# checking unique values from the dataset

for i in data1.columns:
    print({i:data1[i].unique()})
```

```python
# replacing the '?' to nan values in the dataset

data1=data1.replace(' ?',np.nan)
data1
```

```python
# after relacing checking null values

data1.isnull().sum()
```

```python
# filling  the null values by using mode value 

for i in ['workclass','occupation','native_country']:
    data1[i].fillna(data1[i].mode()[0],inplace=True)
```

```python
# again checking null value 

data.isnull().sum()
```

```python
# creating object datatype from the dataset

colname=[]
for i in data1:
    if data1[i].dtype=='object':
        colname.append(i)
colname
```

```python
# converting object data into numerical data

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in colname:
    data1[i]=le.fit_transform(data1[i])       # sklearn function 
    
data1

# income (0----> less the 50k , 1----> greater then 50k) on data remeber
```

```python
x=data1.values[:,0:-1]
y=data1.values[:,-1]
```

```python

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
print(y_pred)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm=confusion_matrix(y_test,y_pred)
print(cfm)

print('classification_report:')
print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test,y_pred)
print('accuracy of thr model :',acc)
```
